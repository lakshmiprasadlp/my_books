You're now working with `exclude_unset=True` in `model_dump()` ‚Äî a great way to **serialize only the fields that were explicitly set by the user**. Let's break this down clearly and deeply.

---

## ‚úÖ Your Code Summary

```python
from pydantic import BaseModel

class Address(BaseModel):
    city: str
    state: str
    pin: str

class Patient(BaseModel):
    name: str
    gender: str = 'Male'  # Default value
    age: int
    address: Address

address_dict = {'city': 'gurgaon', 'state': 'haryana', 'pin': '122001'}
address1 = Address(**address_dict)

patient_dict = {'name': 'nitish', 'age': 35, 'address': address1}
patient1 = Patient(**patient_dict)

temp = patient1.model_dump(exclude_unset=True)

print(temp)
print(type(temp))
```

---

## üîç What does `exclude_unset=True` do?

When used in `.model_dump()`:

* It **excludes fields** that were **not explicitly set** (i.e., those using **default values**).
* Useful for generating **clean, compact JSON**, especially for **partial updates (PATCH requests)** or API responses.

---

## üß™ Breakdown of `Patient`

| Field     | Was It Set in Input? | Value          | Included in Dump?       |
| --------- | -------------------- | -------------- | ----------------------- |
| `name`    | ‚úÖ Yes                | `'nitish'`     | ‚úÖ Yes                   |
| `gender`  | ‚ùå No (default used)  | `'Male'`       | ‚ùå No (excluded)         |
| `age`     | ‚úÖ Yes                | `35`           | ‚úÖ Yes                   |
| `address` | ‚úÖ Yes (object)       | `Address(...)` | ‚úÖ Yes (with all fields) |

---

## üñ® Output

```python
{
    'name': 'nitish',
    'age': 35,
    'address': {
        'city': 'gurgaon',
        'state': 'haryana',
        'pin': '122001'
    }
}
<class 'dict'>
```

‚úÖ Notice that `'gender'` is **excluded** because it's a **default** value and wasn't provided in `patient_dict`.

---

## üîÑ What if You Remove `exclude_unset=True`?

```python
patient1.model_dump()
```

Output will include:

```python
{
    'name': 'nitish',
    'gender': 'Male',  # Now included!
    'age': 35,
    'address': {
        'city': 'gurgaon',
        'state': 'haryana',
        'pin': '122001'
    }
}
```

---

## üß† When to Use `exclude_unset=True`

| Use Case                     | Why Use It                           |
| ---------------------------- | ------------------------------------ |
| API PATCH/partial update     | Avoid sending unchanged/default data |
| Logging/debugging user input | See what user actually provided      |
| Compact serialization        | Reduce payload size                  |

---

## ‚úÖ Summary

| Parameter                    | Effect                                              |
| ---------------------------- | --------------------------------------------------- |
| `exclude_unset=True`         | Excludes fields that weren‚Äôt provided by user input |
| Default fields (like gender) | Will be omitted unless explicitly set               |
| Output Type                  | Still a `dict` (confirmed by `type(temp)`)          |
