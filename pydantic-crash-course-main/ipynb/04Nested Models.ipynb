## ‚úÖ What Are Nested Models?

Nested models are when a field in a `BaseModel` is itself another `BaseModel`.

### Why Use Them?

* üîÅ **Reusability**: Define once, reuse anywhere (like `Address`, `Vitals`, etc.)
* üîí **Automatic Validation**: Nested models are validated **recursively**
* üßº **Clean structure**: Logical grouping of related fields
* üì§ **Serialization support**: Works seamlessly with `.model_dump()`, `.json()` and FastAPI

---

## üì¶ Your Code Explained

### 1. Define the Nested Model

```python
class Address(BaseModel):
    city: str
    state: str
    pin: str
```

---

### 2. Use it inside `Patient`

```python
class Patient(BaseModel):
    name: str
    gender: str
    age: int
    address: Address  # Nested model
```

This means: `address` **must** be an instance of `Address` (or a dict that can become one).

---

### 3. Create Instances

```python
address_dict = {'city': 'gurgaon', 'state': 'haryana', 'pin': '122001'}
address1 = Address(**address_dict)

patient_dict = {'name': 'nitish', 'gender': 'male', 'age': 35, 'address': address1}
patient1 = Patient(**patient_dict)
```

‚úÖ Everything is correctly validated and parsed ‚Äî both top-level and nested fields.

---

### 4. Dumping Only Specific Fields

You're trying to use:

```python
temp = patient1.model_dump(include=)
```

Let‚Äôs now **fix and explore that**.

---

## üõ† How to Use `include` in `.model_dump()`

### ‚úÖ Example 1: Include Only Top-Level Fields

```python
temp = patient1.model_dump(include={"name", "address"})
print(temp)
```

Output:

```python
{
    'name': 'nitish',
    'address': {'city': 'gurgaon', 'state': 'haryana', 'pin': '122001'}
}
```

---

### ‚úÖ Example 2: Include Nested Fields

```python
temp = patient1.model_dump(include={"address": {"city", "pin"}})
print(temp)
```

Output:

```python
{
    'address': {'city': 'gurgaon', 'pin': '122001'}
}
```

You can **filter deeply** into nested models using a **nested dict in `include`**.

---

### ‚úÖ Example 3: Combine Both Levels

```python
temp = patient1.model_dump(include={"name", "address": {"city"}})
print(temp)
```

Output:

```python
{
    'name': 'nitish',
    'address': {'city': 'gurgaon'}
}
```

## üì¶ Extra Benefits of Nested Models

### ‚úÖ Reusability

```python
class Insurance(BaseModel):
    provider: str
    policy_number: str

class MedicalRecord(BaseModel):
    patient: Patient
    insurance: Insurance
```

Now you can use the same `Patient` or `Address` in multiple places!

---

### ‚úÖ Validation

Even if someone gives bad input for the nested address:

```python
patient_dict = {
    'name': 'nitish',
    'gender': 'male',
    'age': 35,
    'address': {'city': 'gurgaon', 'pin': 122001}  # missing 'state'
}

patient1 = Patient(**patient_dict)
```

Pydantic will raise a `ValidationError` ‚Äî no need to write custom checks.
